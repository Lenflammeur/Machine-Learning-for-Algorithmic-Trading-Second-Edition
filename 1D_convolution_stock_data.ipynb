```python
import pandas_datareader as data
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
%matplotlib inline
```

```python
data_source = 'yahoo'
start_date = '2016-01-01'
end_date = '2021-11-30'
Google = data.DataReader('GOOG', data_source, start_date, end_date)
```

```python
type(Google['Close'])
```

```python
px.line(Google['Close'])
```

```python
Google_close_array = np.array([Google['Close']])[0]
```

```python
def gaussian_filter(size,sigma):
    filter_range = np.linspace(-int(size/2),int(size/2),size)
    gaussian_filter = [1 / (sigma * np.sqrt(2*np.pi)) * np.exp(-x**2/(2*sigma**2)) for x in filter_range]
    return gaussian_filter

Google_close_lowpass_filtred = np.convolve(Google_close_array, gaussian_filter(Google_close_array.size, sigma=10), mode='same')
```

```python
Google['Close_filtred'] = Google_close_lowpass_filtred
```

```python
pd.options.plotting.backend = "plotly"
Google.plot(y=['Close', 'Close_filtred'])
```
